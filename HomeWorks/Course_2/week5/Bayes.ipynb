```python
from sklearn import datasets
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.model_selection import cross_val_score
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
digits_X = pd.DataFrame(datasets.load_digits().data)
digits_y = pd.DataFrame(datasets.load_digits().target)
breast_cancer_X = pd.DataFrame(datasets.load_breast_cancer().data)
breast_cancer_y = pd.DataFrame(datasets.load_breast_cancer().target)
```

```python
GNB = GaussianNB()
BNB = BernoulliNB()
MNB = MultinomialNB()
```

```python
print(cross_val_score(GNB, digits_X, digits_y, cv=3).mean())
print(cross_val_score(BNB, digits_X, digits_y, cv=3).mean())
print(cross_val_score(MNB, digits_X, digits_y, cv=3).mean())
```

```python
print(cross_val_score(GNB, breast_cancer_X, breast_cancer_y, cv=3).mean())
print(cross_val_score(BNB, breast_cancer_X, breast_cancer_y, cv=3).mean())
print(cross_val_score(MNB, breast_cancer_X, breast_cancer_y, cv=3).mean())
```
