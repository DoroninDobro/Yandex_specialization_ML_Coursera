```python
%matplotlib inline
from matplotlib import pylab as plt
import numpy as np
from scipy import linalg
from math import sin, exp
from scipy import interpolate  # по известным значениям точек - находим другие, интерполируем
```

```python
def f(i):
    y = (sin(i / 5) * exp(i / 10) + 5 * exp(-i / 2))
    return y
print(f(1), f(4), f(10), f(15))
```

```python
x = np.arange(1, 16, 0.1)
y = []
for i in x:
    y.append(sin(i / 5) * exp(i / 10) + 5 * exp(-i / 2))
y = np.array(y)
plt.plot(x, y)
plt.show()
# this base function
```

```python
a = np.array([[1, 1], [1, 15]])  # matrix
b = np.array([3.25, 0.635]) # solve

x = linalg.solve(a, b) # we search x, who matrix * x = solve
print(x)
```
